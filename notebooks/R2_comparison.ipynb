```python
import pandas as pd
import numpy as np

import seaborn as sns
import pylab as plt
```

```python
# load R2 values
opdc = pd.read_csv(f'/scratch/c.c21013066/data/opdc/analysis/progression/r2.csv').rename(columns={'Unnamed: 0':'clinical target'})
track = pd.read_csv(f'/scratch/c.c21013066/data/opdc/tracking/r2.csv').rename(columns={'Unnamed: 0':'clinical target'})
ppmi = pd.read_csv(f'/scratch/c.c21013066/data/ppmi/phenotypes/phenotypes/r2.csv').rename(columns={'Unnamed: 0':'clinical target'})
```

```python
# data munging to same format and names
ppmi_dic = {'updrs_i':'UPDRS_I','updrs_ii':'UPDRS_II','updrs_iii':'UPDRS_III','updrs_iv':'UPDRS_IV','moca':'MOCA_total','rbd':'RBD_total','epworth':'ESS_total','se_adl':'Schwab_England_ADL',
      'systolic_bp_drop':'SBP_postural_drop'}
track_dic = {'seman_flu_score':'semantic_fluency','Schwab_england':'Schwab_England_ADL'}
opdc_dic = {'seman_flu_score':'semantic_fluency','MOCA_total_adj':'MOCA_total','sniffin_total':'upsit'}

ppmi['clinical target'] = ppmi['clinical target'].replace(ppmi_dic)
track['clinical target'] = track['clinical target'].replace(track_dic)
opdc['clinical target'] = opdc['clinical target'].replace(opdc_dic)
ppmi['cohort'] = 'PPMI'
track['cohort'] = 'Tracking'
opdc['cohort'] = 'Discovery'
```

```python
# merge
merged = pd.concat([ppmi,track,opdc])
merged = merged.pivot(columns=['cohort'],index='clinical target')
merged = merged.swaplevel(axis='columns').sort_index(axis=1)
merged = merged.iloc[merged.isnull().sum(axis=1).mul(-1).argsort()[::-1]]
```

```python
# save
merged.to_csv('/scratch/c.c21013066/data/opdc/cohort_comparison_r2_table.csv')
```
