```python
import pandas as pd
import numpy as np
```

```python
# load raw info for each cohort
table_opdc = pd.read_csv('/scratch/c.c21013066/data/opdc/table.csv',index_col=0).drop(columns=['age_onset']).drop(index=['min','25%','50%','75%','max'])
table_opdc['Cohort'] = 'Discovery'
table_tracking = pd.read_csv('/scratch/c.c21013066/data/opdc/tracking/table.csv',index_col=0).drop(index=['min','25%','50%','75%','max'])
table_tracking['Cohort'] = 'Tracking'
table_ppmi = pd.read_csv('/scratch/c.c21013066/data/ppmi/phenotypes/phenotypes/table.csv',index_col=0).drop(index=['min','25%','50%','75%','max'])
table_ppmi['Cohort'] = 'PPMI'
```

```python
merged = pd.concat([table_ppmi.reset_index(),table_opdc.reset_index(),table_tracking.reset_index()]).rename(columns={'index':'statistic'})
```

```python
table = merged.pivot_table(columns=['Cohort','statistic']).round(2)
table.loc[:,(slice(None),'count')] = table.loc[:,(slice(None),'count')].astype(int)
```
